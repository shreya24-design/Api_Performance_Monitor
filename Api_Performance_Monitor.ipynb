```python
import requests
import time
import pandas as pd

websites = [
    "https://www.google.com",
    "https://github.com",
    "https://openai.com"
]

results = []

for site in websites:
    start = time.time()

    response = requests.get(site)

    end = time.time()

    response_time = round(end - start, 3)

    results.append([site, response.status_code, response_time])

df = pd.DataFrame(
    results,
    columns=["Website", "Status Code", "Response Time (sec)"]
)

print(df)
```

```python
df.to_csv("api_performance_report.csv", index=False)

print("API Performance Report Saved Successfully!")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.bar(df["Website"], df["Response Time (sec)"])

plt.title("API Response Time Monitor")
plt.xlabel("Website")
plt.ylabel("Response Time (sec)")

plt.show()
```
